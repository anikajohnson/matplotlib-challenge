## Observations and Insights

1. Overall the dataset is pretty well distributed with 499 mice split between 10 trials with each drug trial administered to 25 mice each, aside from Stelasyn which is missing one mouse. There was likely an issue with this mouse for some reason and its trail data was thrown out. In addition, the dataset is evenly distributed between male and female mice with 50.6% male and 49.4% female, a difference of 0.6%.

2. Capomulin and Ramicane represent the best treatment options with the lowest average (mean) tumor volume sizes at 40.68 and 40.22 respectively as well as the lowest standard deviations. Looking at the top four treatment options, Ramicane and Capomulin show a tight distribution of values within the inner quartile range (within the 25th and 75th percentiles) while Ceftamin has a much large IQR and Infubinol shows an outlier value.  

3. Average mouse weight and average tumor volume are positively correlated with a r-squared value of 0.84, meaning that 84% of the variability of data around the mean is explained by the model. This may indicate that at higher weights the mice generally have more tissue that is susceptible to tumor growth. 


## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

mouse_metadata
```

```python
study_results
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")
combined_data
```

## Summary statistics

```python
#Generate a summary statistics table consisting of the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen

#Calculate mean
mean_volume = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)'],2))
mean_volume = mean_volume.rename(columns={"Tumor Volume (mm3)":"Mean"})
mean_volume
```

```python
#Calculate median
median_volume = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)'],2))
median_volume = median_volume.rename(columns={"Tumor Volume (mm3)":"Median"})
median_volume
```

```python
#Calculate variance
variance = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)'],2))
variance = variance.rename(columns={"Tumor Volume (mm3)":"Variance"})
variance
```

```python
#Calculate standard deviation
stdev = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)'],2))
stdev = stdev.rename(columns={"Tumor Volume (mm3)":"StDev"})
stdev
```

```python
#Calculate standard error
SEM = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)'],2))
SEM = SEM.rename(columns={"Tumor Volume (mm3)":"SEM"})
SEM
```

```python
summary_stat = [mean_volume, median_volume, variance, stdev, SEM]
summary_stat = pd.concat(summary_stat,axis=1,join='inner')
summary_stat
```

```python
stat_analysis= pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).describe()['Tumor Volume (mm3)'],2))
stat_analysis
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the 
#most promising treatment regimens. 
#Calculate the IQR and quantitatively determine if there are any 
#potential outliers. 

top_treatments = combined_data.loc[((combined_data["Drug Regimen"]=="Ramicane") |
                                        (combined_data["Drug Regimen"]=="Capomulin") |
                                         (combined_data["Drug Regimen"]=="Infubinol") |
                                        (combined_data["Drug Regimen"]=="Ceftamin"))]

top_treatments_grouped = top_treatments.groupby(["Mouse ID", "Drug Regimen"])["Timepoint"].max()

top_treatments_grouped_df = pd.DataFrame(top_treatments_grouped)


top_treatments_merged_df = pd.merge(top_treatments, top_treatments_grouped_df, how ="inner", on =["Mouse ID", "Drug Regimen", "Timepoint"])


ramicane_final_vol = top_treatments_merged_df.loc[top_treatments_merged_df["Drug Regimen"]=="Ramicane", 
                                                ["Tumor Volume (mm3)"]]
capomulin_final_vol = top_treatments_merged_df.loc[top_treatments_merged_df["Drug Regimen"]=="Capomulin", 
                                                 ["Tumor Volume (mm3)"]]
infubinol_final_vol = top_treatments_merged_df.loc[top_treatments_merged_df["Drug Regimen"]=="Infubinol", 
                                                 ["Tumor Volume (mm3)"]]
ceftamin_final_vol = top_treatments_merged_df.loc[top_treatments_merged_df["Drug Regimen"]=="Ceftamin", 
                                                ["Tumor Volume (mm3)"]]

```

```python
ram_quartiles = ramicane_final_vol["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
ram_lower_q = ram_quartiles[0.25]
ram_upper_q = ram_quartiles[0.75]
ram_iqr = ram_upper_q-ram_lower_q

ram_lower_bound = ram_lower_q - (1.5 * ram_iqr)
ram_upper_bound = ram_upper_q + (1.5 * ram_iqr)

print(f"Values below {ram_lower_bound} could be outliers for Ramicane.")
print(f"Values above {ram_upper_bound} could be outliers for Ramicane.")
```

```python
cap_quartiles = capomulin_final_vol["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
cap_lower_q = cap_quartiles[0.25]
cap_upper_q = cap_quartiles[0.75]
cap_iqr = cap_upper_q-cap_lower_q

cap_lower_bound = cap_lower_q - (1.5 * cap_iqr)
cap_upper_bound = cap_upper_q + (1.5 * cap_iqr)

print(f"Values below {cap_lower_bound} could be outliers for Capomulin.")
print(f"Values above {cap_upper_bound} could be outliers for Capomulin.")
```

```python
inf_quartiles = infubinol_final_vol["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
inf_lower_q = inf_quartiles[0.25]
inf_upper_q = inf_quartiles[0.75]
inf_iqr = inf_upper_q-inf_lower_q

inf_lower_bound = inf_lower_q - (1.5 * inf_iqr)
inf_upper_bound = inf_upper_q + (1.5 * inf_iqr)

print(f"Values below {inf_lower_bound} could be outliers for Infubinol.")
print(f"Values above {inf_upper_bound} could be outliers for Infubinol.")
```

```python
cef_quartiles = ceftamin_final_vol["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
cef_lower_q = cef_quartiles[0.25]
cef_upper_q = cef_quartiles[0.75]
cef_iqr = cef_upper_q-cef_lower_q

cef_lower_bound = cef_lower_q - (1.5 * cef_iqr)
cef_upper_bound = cef_upper_q + (1.5 * cef_iqr)

print(f"Values below {cef_lower_bound} could be outliers for Ceftamin.")
print(f"Values above {cef_upper_bound} could be outliers for Ceftamin.")
```

```python
# Generate a box plot of the final tumor volume of each mouse 
#across four regimens of interest

fig, axs = plt.subplots(1,4)
fig.suptitle("Top Treatments")


axs[0].boxplot(ramicane_final_vol["Tumor Volume (mm3)"])
axs[0].set_title("Ramicane")
axs[0].set_ylabel("Final Tumor Volume (mm3)")

axs[1].boxplot(capomulin_final_vol["Tumor Volume (mm3)"])
axs[1].set_title("Capomulin")
axs[1].set_ylabel("Final Tumor Volume (mm3)")

axs[2].boxplot(infubinol_final_vol["Tumor Volume (mm3)"])
axs[2].set_title("Infubinol")
axs[2].set_ylabel("Final Tumor Volume (mm3)")

axs[3].boxplot(ceftamin_final_vol["Tumor Volume (mm3)"])
axs[3].set_title("Ceftamin")
axs[3].set_ylabel("Final Tumor Volume (mm3)")


plt.subplots_adjust(left = 0.1, bottom = None, right = 2.75, top = None, wspace = 0.2, hspace = 0.2)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume 
#for the Capomulin regimen

capomulin_data = combined_data.loc[combined_data["Drug Regimen"] == "Capomulin"]

capomulin_data_grouped = capomulin_data.groupby(["Mouse ID"])
mouse_weight = capomulin_data_grouped["Weight (g)"].mean()
mouse_tumor_vol = capomulin_data_grouped["Tumor Volume (mm3)"].mean()

plt.scatter(mouse_weight, mouse_tumor_vol)

plt.title("Average Mouse Weight & Tumor Volume for Mice in Capomulin Trial")
plt.xlabel("Average Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")


plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model
#for mouse weight and average tumor volume for the Capomulin regimen
slope, intercept, rvalue, pvalue, stderr = st.linregress(mouse_weight, mouse_tumor_vol)
regress_values = mouse_weight* slope + intercept
lineEq = "y = " + str(round(slope,2)) + "x + " +str(round(intercept,2))

plt.scatter(mouse_weight, mouse_tumor_vol)
plt.plot(mouse_weight, regress_values, color ="red")
plt.annotate(lineEq, (20, 36), fontsize = 15, color = "red")

plt.title("Average Mouse Weight & Tumor Volume for Mice in Capomulin Trial")
plt.xlabel("Average Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The r-squared is: {rvalue}")
plt.show()
```
